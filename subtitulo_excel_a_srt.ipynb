```python
import pandas as pd
from datetime import timedelta, datetime

# *************** function ***************
def add_time(hour, min, secs, add_hs=0, add_min=0, add_sec=0, add_msec=0):
    d = hour + ':' + min + ':' + secs
    dt = datetime.strptime(d, '%H:%M:%S.%f')
    result = dt + timedelta(hours   = add_hs,
                            minutes = add_min,
                            seconds = add_sec,
                            milliseconds = add_msec)
    letter = ("%02d:%02d:%02d,%s" % (result.hour, result.minute, result.second, str(result.microsecond)[:2]))
    return letter


def srt_file_write_plus(srt_file, 
                        xlsx_df, 
                        add_hs   = 0, 
                        add_min  = 0,
                        add_sec  = -34,
                        add_msec = 0,
                        sec_linea_final = 30,
                    ):

    file_object  = open(srt_file, "w+")

    line = ""
    subtitle_esp = ""
    for i in range(0,len(xlsx_df)):
        if(i == len(xlsx_df)-1):
            line = str(int(xlsx_df["연번"][i])) + "\n"
            time_line = xlsx_df["타임코드(In)"][i].split(":")
            time_line_added = add_time(time_line[0],time_line[1],time_line[2],
                                        add_hs   = add_hs ,
                                        add_min  = add_min,
                                        add_sec  = add_sec,
                                        add_msec = add_msec)
            line+= time_line_added + " --> "
            time_line_added = add_time(time_line[0],time_line[1],time_line[2],
                                        add_hs   = add_hs ,
                                        add_min  = add_min,
                                        add_sec  = add_sec+sec_linea_final,
                                        add_msec = add_msec)
            line+= time_line_added + "\n"
            line+= str(xlsx_df["번역문"][i]) + "\n"
            # print(line)
            file_object.write(line)
            break

        if(pd.isna(xlsx_df["타임코드(In)"][i])):
            subtitle_esp += str(xlsx_df["번역문"][i]) + "\n"
        else:
            line = str(int(xlsx_df["연번"][i])) + "\n"
            time_line = xlsx_df["타임코드(In)"][i].split(":")
            time_line_added = add_time(time_line[0],time_line[1],time_line[2],
                                        add_hs   = add_hs ,
                                        add_min  = add_min,
                                        add_sec  = add_sec,
                                        add_msec = add_msec)
            line+= time_line_added + " --> "
            subtitle_esp = str(xlsx_df["번역문"][i]) + "\n"
        if(pd.isna(xlsx_df["타임코드(In)"][i+1])==False):
            time_line = xlsx_df["타임코드(In)"][i+1].split(":")
            time_line_added = add_time(time_line[0],time_line[1],time_line[2],
                                        add_hs   = add_hs ,
                                        add_min  = add_min,
                                        add_sec  = add_sec,
                                        add_msec = add_msec)
            line+= time_line_added + "\n"
            line+= subtitle_esp+"\n"
            file_object.write(line)

    file_object.close()

# *************** main ***************
excel_file = "sub_excel.xlsx"
srt_file = "sub_srt.srt"

row_start = 26
row_end = 972
add_hs   = 0 
add_min  = 0
add_sec  = -34
add_msec = 0
sec_linea_final = 30

specific_rows = list(range(26,972))
usecols = "B:F"


end_transformation = False
while end_transformation==False:
    xlsx_df = pd.read_excel(excel_file, usecols="B:F", skiprows = lambda x: x not in specific_rows)
    srt_file_write_plus(srt_file, xlsx_df, add_hs, add_min, add_sec, add_msec, sec_linea_final)
    continuar = input("Una vez más? (si/no)")
    while (continuar == "si" or continuar == "no")==False:
        continuar = input("Ingrese si o no.\nQuiere generar .srt vez más? (si/no)")
    if(continuar == "no"):
        end_transformation = True
```
